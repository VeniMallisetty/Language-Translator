```python
# This is the final code!
from googletrans import Translator

def get_user_input():
    input_text = input("Enter the text in English: ")
    return input_text

def get_target_language():
    # Display available languages
    print("\nAvailable Languages:")
    for code, language in languages.items():
        print(f"{code}: {language}")

    target_language = input("Enter the target language code (e.g., 'es' for Spanish, 'fr' for French): ")
    return target_language

def translate_text(input_text, target_language):
    translator = Translator()
    translated_text = translator.translate(input_text, dest=target_language)
    return translated_text.text

if __name__ == "__main__":
    # Display all language options
    print("Available Languages:")
    for code, language in languages.items():
        print(f"{code}: {language}")

    # Get user input
    input_text = get_user_input()

    # Get target language
    target_language = get_target_language()

    # Perform translation
    translated_text = translate_text(input_text, target_language)

    # Print the results
    print(f"\nInput Text: {input_text}")
    print(f"Translated Text ({target_language}): {translated_text}")
```
